```python
import pandas as  pd
import numpy as np 
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.describe()
```

```python
print (df['tip'].mean()," ",
df['tip'].min()," ",
df['tip'].max()," ",
df['tip'].median()," ",
df['tip'].std()," ")
```

```python
summary_stats = df.groupby('sex').agg({
    'total_bill': ['mean', 'median', 'min', 'max', 'std'],
    'tip': ['mean', 'median', 'min', 'max', 'std']
})
summary_stats
```
